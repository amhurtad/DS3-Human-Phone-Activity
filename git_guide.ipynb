## Usage: Actually Working with it

1. Make changes to code/files/whatever you need to do
2. Save changes
3. Open up git bash
4. Type following in order:
- git add .
- git commit -m "message that explains your changes"
- git pull
- git push

## Git commands/explanations
1. git add filename
- tells git to track the given file, having . as the filename tracks everything in the folder
2. git commit
- creates a new commit instance(locally) containing the current contents of the index
- can be followed by alot of things, but -m is the only one we really need and it stands for message
3. git pull
- pulls the most recent commit from the online repository, updating your local files
4. git push
- pushes your most recent commit instance to the online repository
5.  git config --global user.name "FIRST_NAME LAST_NAME" AND git config --global user.email "MY_NAME@example.com"
- tells your git who you are, is an authentication system
- might see an error telling you to use these commands on the first time running git commands
6. git log
- shows you your commit history
7. git status
- shows you which files are tracked/untracked by git
8. git clone repositoryLink
- clones the repository, allowing you to get the code/work on it (if you have the permissions)
