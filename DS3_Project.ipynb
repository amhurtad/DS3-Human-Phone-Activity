# Exploratory Data Analysis

```python
#TSNE is used for creating visuals on high dimensional data
from sklearn.manifold import TSNE
```

```python
#only taking the columns that will be independent variables
scatter_x = df_training.iloc[:,:-2]
```

```python
scatter_x.head()
```

```python
#Fitting to our training set
tsne = TSNE(perplexity = 50).fit_transform(scatter_x)
```

```python
plt.figure(figsize = (15,10))
sns.scatterplot(x = tsne[:,0], y = tsne[:,1], hue = df_training["Activity"], palette = 'dark')
```

```python
#Make a boxplot with one specific feature
df_training.columns.to_list()
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(x = 'Activity', y = 'tBodyAccJerkMag-mean()', data = df_training, palette = 'colorblind', showfliers = False)
plt.title("Body Acceleration Jerk Mean Across Activities")
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(x = 'Activity', y = 'tGravityAccMag-mean()', data = df_training, palette = 'colorblind', showfliers = False)
plt.title("Gravity Acceleration Jerk Mean Across Activities")
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(x = 'Activity', y = 'tBodyAccMag-mean()', data = df_training, palette = 'colorblind', showfliers = False)
plt.title("Body Acceleration Mean Across Activities")
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(x = 'Activity', y = 'tBodyGyroMag-mean()', data = df_training, palette = 'colorblind', showfliers = False)
plt.title("Body Gyro Mean Across Activities")
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(x = 'Activity', y = 'angle(X,gravityMean)', data = df_training, palette = 'colorblind', showfliers = False)
plt.title("Angle Gravity Mean Across Activities")
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(x = 'Activity', y = 'fBodyAccMag-energy()', data = df_training, palette = 'colorblind', showfliers = False)
plt.title("Angle Gravity Mean Across Activities")
plt.show()
```

```python
#Looking at Gyro mean across activities through line plot
plt.figure(figsize = (15,10))
sns.distplot(df_training[df_training['Activity'] == 'SITTING']['tBodyGyroMag-mean()'], hist = False, label = 'Sitting')
sns.distplot(df_training[df_training['Activity'] == 'LAYING']['tBodyGyroMag-mean()'], hist = False, label = 'Laying')
sns.distplot(df_training[df_training['Activity'] == 'STANDING']['tBodyGyroMag-mean()'], hist = False, label = 'Standing')
sns.distplot(df_training[df_training['Activity'] == 'WALKING_UPSTAIRS']['tBodyGyroMag-mean()'], hist = False, label = 'Walking Upstairs')
sns.distplot(df_training[df_training['Activity'] == 'WALKING_DOWNSTAIRS']['tBodyGyroMag-mean()'], hist = False, label = 'Walking Downstairs')
sns.distplot(df_training[df_training['Activity'] == 'WALKING']['tBodyGyroMag-mean()'], hist = False, label = 'Walking')
plt.title('Body Gyro Mean across Activites')
plt.show()
```

```python
x = df_training.get("tBodyAcc-mean()-X")
y = df_training.get("tBodyAcc-mean()-Y")
z = df_training.get("tBodyAcc-mean()-Z")
activity = df_training.get("Activity")
```

```python
new_df = pd.DataFrame().assign(
    x = x,
    y = y,
    z = z,
    activity = activity
)
```

```python
new_df = new_df[new_df['activity'].isin(["STANDING", "SITTING", "LAYING"])]
only_standing = new_df[new_df.get("activity") == "STANDING"]
only_sitting = new_df[new_df.get("activity") == "SITTING"]
only_laying = new_df[new_df.get("activity") == "LAYING"]
```

```python
ax = plt.axes(projection='3d')
ax.scatter3D(only_standing.get('x'), only_standing.get('y'), only_standing.get('z'), 
             c = "red", label = "STANDING")

ax.scatter3D(only_sitting.get('x'), only_sitting.get('y'), only_sitting.get('z'), 
             c = "blue", label = "SITTING")

ax.scatter3D(only_laying.get('x'), only_laying.get('y'), only_laying.get('z'), 
             c = "green", label = "LAYING", alpha = 0.5)

ax.legend()
plt.show()
```
